```python
#Load data 

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../ml/leads.csv")
df.head()
```

```python
#Conversion Rate Overall

df ["converted"].mean()
```

```python
#Budget vs Conversion

df.groupby("budget")["converted"].mean().plot(kind="bar", title="Conversion Rate by Budget")
plt.show()
```

```python
#Urgency Buckets

df["urgency_bucket"] = pd.cut(df["urgency"], bins=[0, 30, 60, 90], labels=["0-30", "31-60", "61-90"])
df.groupby("urgency_bucket")["converted"].mean().plot(kind="bar", title="Conversion by Urgency")
plt.show()
```

```python
#Preapproval Impact

df.groupby("preapproved")["converted"].mean().plot(kind="bar", title="Conversion by Mortgage Preapproval")
plt.show()
```
